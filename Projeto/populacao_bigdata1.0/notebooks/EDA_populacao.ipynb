# EDA População – Notebook inicial
Este notebook lê dados `silver`/`gold` (quando existirem) e faz alguns gráficos.

```python
import pandas as pd
import plotly.express as px
from pathlib import Path

base = Path('..')
silver = base/'data'/'silver'/'population_clean.parquet'
bronze_csv = base/'data'/'raw'/'worldbank_population.csv'

try:
    df = pd.read_parquet(silver)
except Exception:
    df = pd.read_csv(bronze_csv)

df.head()
```

## Evolução populacional por país

```python
fig = px.line(df, x='year', y='population', color='country_name', markers=True, title='População por país ao longo do tempo')
fig.show()
```
